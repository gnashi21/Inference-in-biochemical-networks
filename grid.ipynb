```python
import matplotlib.pyplot as plt
import numpy as np
directories = ['hist 1', 'hist 2', 'hist 3', 'hist 4', 'hist 5', 'hist 6', 'hist 7', 'hist 8', 'hist 9']
dts = ['0.05', '0.1', '0.2']
points = [100, 1000, 10000]
fig, axes = plt.subplots(3, 3, figsize=(12, 12))

row1 = directories[:3]
row2 = directories[3:6]
row3 = directories[6:]

rows = [row1, row2, row3]

for i, row in enumerate(rows):
    for j, directory in enumerate(row):
        dt = dts[j] 
        point = points[i]

        # Load histogram image
        filename = f'{directory}/alpha_histogram.png'
        img = plt.imread(filename)

        # Plot histogram
        axes[i, j].imshow(img)
        axes[i, j].axis('off')
        if j == 0:
            axes[i, j].set_ylabel(f'Points: {point}')
        if i == 0:
            axes[i, j].set_title(f'dt = {dt}')

       
        axes[i, j].text(0.5, -0.1, directory, ha='center', va='center', transform=axes[i, j].transAxes)

plt.tight_layout()
plt.show()
```
